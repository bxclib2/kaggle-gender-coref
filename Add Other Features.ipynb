```python
import pandas as pd
```

```python
gap_train= pd.read_pickle('./temp_result/train_kaggle_processed_PPA_PCA_PPA_neither.p')
gap_test= pd.read_pickle('./temp_result/test_kaggle_processed_PPA_PCA_PPA_neither.p')
gap_valid= pd.read_pickle('./temp_result/valid_kaggle_processed_PPA_PCA_PPA_neither.p')
```

```python
gap_train.head()
```

```python
def is_topic(name, topic):
    for i in name:
        if i in topic:
            return 1
    return 0
```

```python
gap_train['topic'] = gap_train['URL'].apply(lambda x: x.split('/')[-1].split('_'))
gap_valid['topic'] = gap_valid['URL'].apply(lambda x: x.split('/')[-1].split('_'))
gap_test['topic'] = gap_test['URL'].apply(lambda x: x.split('/')[-1].split('_'))
```

```python
gap_train['topic_A'] = gap_train.apply(lambda x: is_topic(x['A'].split(' '), x['topic']), axis=1)
gap_test['topic_A'] = gap_test.apply(lambda x: is_topic(x['A'].split(' '), x['topic']), axis=1)
gap_valid['topic_A'] = gap_valid.apply(lambda x: is_topic(x['A'].split(' '), x['topic']), axis=1)

gap_train['topic_B'] = gap_train.apply(lambda x: is_topic(x['B'].split(' '), x['topic']), axis=1)
gap_test['topic_B'] = gap_test.apply(lambda x: is_topic(x['B'].split(' '), x['topic']), axis=1)
gap_valid['topic_B'] = gap_valid.apply(lambda x: is_topic(x['B'].split(' '), x['topic']), axis=1)
```

```python
gap_train.to_pickle('./temp_result/train_kaggle_processed_PPA_PCA_PPA_neither_topic.p')
gap_test.to_pickle('./temp_result/test_kaggle_processed_PPA_PCA_PPA_neither_topic.p')
gap_valid.to_pickle('./temp_result/valid_kaggle_processed_PPA_PCA_PPA_neither_topic.p')
```
